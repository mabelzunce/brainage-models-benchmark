```python
ENIGMA_UNSAM.rename(columns={'Absolute_Error': 'Absolute Error', 'Edad':'Age'}, inplace=True)
ENIGMA_UNSAM=ENIGMA_UNSAM[['ID','Absolute Error','Error','Age']]

ENIGMA_ADNI=ENIGMA_ADNI[['ID','Absolute Error','Error','Age']]

ENIGMA_JUK.rename(columns={'participant_id': 'ID', 'age':'Age'}, inplace=True)
ENIGMA_JUK=ENIGMA_JUK[['ID','Absolute Error','Error','Age']]
ENIGMA_JUK['ID']='JUK_' + ENIGMA_JUK['ID']

ENIGMA_RRIB.rename(columns={'participant_id': 'ID', 'age':'Age'}, inplace=True)
ENIGMA_RRIB=ENIGMA_RRIB[['ID','Absolute Error','Error','Age']]

concat_ENIGMA = pd.concat([ENIGMA_UNSAM, ENIGMA_ADNI, ENIGMA_JUK, ENIGMA_RRIB], ignore_index=True)
```

```python
DBN_UNSAM.rename(columns={'Edad': 'Age'}, inplace=True)
DBN_UNSAM=DBN_UNSAM[['ID','Absolute_Error','Error','Age']]

DBN_ADNI.rename(columns={'Absolute_error': 'Absolute_Error'}, inplace=True)
DBN_ADNI=DBN_ADNI[['ID','Absolute_Error','Error','Age']]

DBN_JUK.rename(columns={'participant_id': 'ID','age':'Age'}, inplace=True)
DBN_JUK.rename(columns={'Absolute_error': 'Absolute_Error'}, inplace=True)
DBN_JUK=DBN_JUK[['ID','Absolute_Error','Error','Age']]
DBN_JUK['ID']='JUK_' + DBN_JUK['ID']

DBN_RRIB.rename(columns={'participant_id': 'ID'}, inplace=True)
DBN_RRIB.rename(columns={'Absolute_error': 'Absolute_Error','age':'Age'}, inplace=True)
DBN_RRIB=DBN_RRIB[['ID','Absolute_Error','Error','Age']]

concat_DBN = pd.concat([DBN_UNSAM, DBN_ADNI, DBN_JUK, DBN_RRIB], ignore_index=True)
```

```python
pyment_UNSAM.rename(columns={'Edad':'Age'},inplace=True)
pyment_UNSAM=pyment_UNSAM[['ID','Absolute_Error','Error','Age']]

pyment_ADNI=pyment_ADNI[['ID','Absolute_Error','Error','Age']]

pyment_JUK.rename(columns={'participant_id': 'ID','age':'Age'}, inplace=True)
pyment_JUK=pyment_JUK[['ID','Absolute_Error','Error','Age']]
pyment_JUK['ID']='JUK_' + pyment_JUK['ID']

pyment_RRIB.rename(columns={'participant_id': 'ID','age':'Age'}, inplace=True)
pyment_RRIB=pyment_RRIB[['ID','Absolute_Error','Error','Age']]

concat_pyment = pd.concat([pyment_UNSAM, pyment_ADNI, pyment_JUK, pyment_RRIB], ignore_index=True)
```

```python
BAN_UNSAM.rename(columns={'Edad':'Age'}, inplace=True)
BAN_UNSAM=BAN_UNSAM[['ID','Absolute_Error','Error','Age']]

BAN_ADNI=BAN_ADNI[['ID','Absolute_Error','Error','Age']]

BAN_RRIB.rename(columns={'participant_id': 'ID','age':'Age'}, inplace=True)
BAN_RRIB=BAN_RRIB[['ID','Absolute_Error','Error','Age']]
BAN_RRIB['ID']='RRIB_' + BAN_RRIB['ID']

BAN_JUK.rename(columns={'participant_id': 'ID','age':'Age'}, inplace=True)
BAN_JUK=BAN_JUK[['ID','Absolute_Error','Error','Age']]

concat_BAN = pd.concat([BAN_UNSAM, BAN_ADNI, BAN_JUK, BAN_RRIB], ignore_index=True)
```

```python
# Mean error for every age bin and model
bins = [20, 30, 40, 50, 60, 70, 80, 90]
labels = ['20-29', '30-39', '40-49', '50-59', '60-69', '70-79', '80-89']

# create a new column 'Age_bin' in each dataframe
concat_BAN['Age_bin'] = pd.cut(concat_BAN['Age'], bins=bins, labels=labels, right=False)
concat_DBN['Age_bin'] = pd.cut(concat_DBN['Age'], bins=bins, labels=labels, right=False)
concat_pyment['Age_bin'] = pd.cut(concat_pyment['Age'], bins=bins, labels=labels, right=False)
concat_ENIGMA['Age_bin'] = pd.cut(concat_ENIGMA['Age'], bins=bins, labels=labels, right=False)
```

```python
mean_error_by_bin_BAN = concat_BAN.groupby('Age_bin')['Error'].mean()
mean_error_by_bin_DBN = concat_DBN.groupby('Age_bin')['Error'].mean()
mean_error_by_bin_pyment = concat_pyment.groupby('Age_bin')['Error'].mean()
mean_error_by_bin_ENIGMA = concat_ENIGMA.groupby('Age_bin')['Error'].mean()
```
