```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
BAN_UNSAM=pd.read_csv('./BrainAgeNeXt/BAN_UNSAM.csv')
BAN_ADNI=pd.read_csv('./BrainAgeNeXt/BAN_ADNI.csv')
BAN_RRIB=pd.read_csv('./BrainAgeNeXt/BAN_RRIB.csv')
BAN_JUK=pd.read_csv('./BrainAgeNeXt/BAN_JUK.csv')
```

```python
fig, axes = plt.subplots(2, 2, figsize=(16, 12))
bin_edges = np.arange(0, 101, 5)

# Subplot 1: UNSAM_LC
sns.histplot(data=BAN_UNSAM, x='Edad', ax=axes[0][1], color='blue', kde=True,bins=bin_edges)
axes[0][1].set_title('UNSAM_LC', fontsize=22)
axes[0][1].set_xlabel('')
axes[0][1].set_ylabel('')
axes[0][1].set_xlim(0, 100)

# Subplot 2: JUK
sns.histplot(data=BAN_JUK, x='age', ax=axes[1][1], color='green', kde=True,bins=bin_edges)
axes[1][1].set_title('JUK', fontsize=22)
axes[1][1].set_xlabel('Age', fontsize=20)
axes[1][1].set_ylabel('Frequency', fontsize=20)
axes[1][1].set_xlim(0, 100)
axes[1][1].set_ylabel('')

# Subplot 3: RRIB
sns.histplot(data=BAN_RRIB, x='age', ax=axes[1][0], color='orange', kde=True,bins=bin_edges)
axes[1][0].set_title('RRIB', fontsize=22)
axes[1][0].set_xlabel('Age', fontsize=20)
axes[1][0].set_ylabel('Frequency', fontsize=20)
axes[1][0].set_xlim(0, 100)



# Subplot 4: ADNI
sns.histplot(data=BAN_ADNI, x='Age', ax=axes[0][0], color='red', kde=True,bins=bin_edges)
axes[0][0].set_title('ADNI', fontsize=22)
axes[0][0].set_xlabel('')
axes[0][0].set_ylabel('Frequency', fontsize=20)
axes[0][0].set_xlim(0, 100)

for ax_row in axes:
    for ax in ax_row:
        ax.tick_params(axis='both', labelsize=16)  # Cambia 16 por el tamao que prefieras

# Ajustar diseo
plt.tight_layout(rect=[0, 0, 1, 0.95])  # deja espacio para el ttulo
plt.show()
```
