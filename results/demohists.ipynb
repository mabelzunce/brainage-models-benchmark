```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
BAN_CP=pd.read_csv('/data/Lautaro/Documentos/BrainAgeCOVID/DATOS/predictions/BrainAgeNeXt/BAN_CP.csv')
BAN_ADNI=pd.read_csv('/data/Lautaro/Documentos/BrainAgeCOVID/DATOS/predictions/BrainAgeNeXt/BAN_ADNI.csv')
BAN_CogConVar=pd.read_csv('/data/Lautaro/Documentos/BrainAgeCOVID/DATOS/predictions/BrainAgeNeXt/BAN_CogConVar.csv')
BAN_NCN=pd.read_csv('/data/Lautaro/Documentos/BrainAgeCOVID/DATOS/predictions/BrainAgeNeXt/BAN_NCN.csv')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Crear figura con 4 subplots (2 filas x 2 columnas)
fig, axes = plt.subplots(1, 4, figsize=(20, 5))
#fig.suptitle('Real Age Distribution Across Cohorts', fontsize=18)

# Subplot 1: CP
sns.histplot(data=BAN_CP, x='Edad', ax=axes[1], color='blue', kde=True)
axes[1].set_title('UNSAM_LC')
axes[1].set_xlabel('Age')
axes[1].set_ylabel('Frequency')
axes[1].set_xlim(0, 100)
axes[1].set_ylabel('')

# Subplot 2: NCN
sns.histplot(data=BAN_NCN, x='age', ax=axes[3], color='green', kde=True)
axes[3].set_title('JUK')
axes[3].set_xlabel('Age')
axes[3].set_ylabel('Frequency')
axes[3].set_xlim(0, 100)
axes[3].set_ylabel('')

# Subplot 3: CogConVar
sns.histplot(data=BAN_CogConVar, x='age', ax=axes[2], color='orange', kde=True)
axes[2].set_title('RRIB')
axes[2].set_xlabel('Age')
axes[2].set_ylabel('Frequency')
axes[2].set_xlim(0, 100)
axes[2].set_ylabel('')


# Subplot 4: ADNI
sns.histplot(data=BAN_ADNI, x='Age', ax=axes[0], color='red', kde=True)
axes[0].set_title('ADNI')
axes[0].set_xlabel('Age')
axes[0].set_ylabel('Frequency')
axes[0].set_xlim(0, 100)
axes[0].set_xlabel('Age')

# Ajustar diseño
plt.tight_layout(rect=[0, 0, 1, 0.95])  # deja espacio para el título
plt.show()
```
