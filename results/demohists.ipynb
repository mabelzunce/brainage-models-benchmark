```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
BAN_CP=pd.read_csv('/data/Lautaro/Documentos/BrainAgeCOVID/DATOS/predictions/BrainAgeNeXt/BAN_CP.csv')
BAN_ADNI=pd.read_csv('/data/Lautaro/Documentos/BrainAgeCOVID/DATOS/predictions/BrainAgeNeXt/BAN_ADNI.csv')
BAN_CogConVar=pd.read_csv('/data/Lautaro/Documentos/BrainAgeCOVID/DATOS/predictions/BrainAgeNeXt/BAN_CogConVar.csv')
BAN_NCN=pd.read_csv('/data/Lautaro/Documentos/BrainAgeCOVID/DATOS/predictions/BrainAgeNeXt/BAN_NCN.csv')
```

```python
# Crear figura y un solo eje
fig, ax = plt.subplots(figsize=(8, 6))
fig.suptitle('Real age distribution', fontsize=16)

# Histogramas
sns.histplot(data=BAN_CP, x='Edad', ax=ax, color='blue', label='CP', kde=True)
sns.histplot(data=BAN_ADNI[BAN_ADNI['Group'] == 'CN'], x='Age', ax=ax, color='red', label='ADNI CN', kde=True)
sns.histplot(data=BAN_ADNI[BAN_ADNI['Group'] != 'CN'], x='Age', ax=ax, color='magenta', label='ADNI clinical', kde=True)
sns.histplot(data=BAN_NCN, x='age', ax=ax, color='green', label='NCN', kde=True)
sns.histplot(data=BAN_CogConVar, x='age', ax=ax, color='orange', label='CogConVar', kde=True)

# Etiquetas y leyenda
ax.set_xlabel('Real age')
ax.set_ylabel('Frequency')
plt.grid(True)
#ax.set_title('Distribucin de Edades en Distintas Cohortes (BAN)')
ax.legend()

plt.tight_layout()
plt.show()
```
