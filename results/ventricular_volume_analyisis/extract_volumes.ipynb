```python
import pandas as pd
import os
import csv
```

```python
ADNI_stats_folder = "/Users/parri/OneDrive/Documentos/Beca PEFI/brainage-models-benchmark/data/recon-all_stats/ADNI"
```

```python
def extract_volumes_aseg(subjects_dir):
    vols_lista = []

    for subj in os.listdir(subjects_dir):
        subj_path = os.path.join(subjects_dir, subj)
        aseg_path = os.path.join(subj_path, "stats", "aseg.stats")

        if not os.path.isdir(subj_path):
            continue
        if not os.path.exists(aseg_path):
            continue

        vols = {
            'SubjID': subj,
            'CortexVol': None,
            'CerebralWhiteMatterVol': None,
            'SubCortGrayVol': None,
            'VentricleChoroidVol': None
        }

        with open(aseg_path, 'r') as f:
            for line in f:
                if line.startswith('# Measure'):
                    partes = line.strip().split(',')

                    if len(partes) >= 4:
                        nombre = partes[1].strip()
                        try:
                            valor = float(partes[3].strip())
                        except ValueError:
                            continue

                        if nombre in vols:
                            vols[nombre] = valor

        vols_lista.append(vols)

    df = pd.DataFrame(vols_lista)
    df.rename(columns={'SubjID': 'SubjectID',
                       'CortexVol':'GMV',
                       'CerebralWhiteMatterVol':'WMV',
                       'SubCortGrayVol':'sGMV',
                       'VentricleChoroidVol':'Ventricles'}, inplace=True)
    return df
```

```python
df=extract_volumes_aseg(ADNI_stats_folder)
df
```

```python
df.to_csv('volumes_asegADNI.csv', index=False)
```
