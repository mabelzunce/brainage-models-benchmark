```python
import os
import pandas as pd

def extraer_volumenes_ventriculares(path_aseg_stats):
    estructuras_ventriculares = [
        'Left-Lateral-Ventricle',
        'Right-Lateral-Ventricle',
        'Left-Inf-Lat-Vent',
        'Right-Inf-Lat-Vent',
        '3rd-Ventricle',
        '4th-Ventricle',
        'Left-choroid-plexus',
        'Right-choroid-plexus'
    ]

    volumen_total = 0.0

    with open(path_aseg_stats, 'r') as f:
        for line in f:
            if line.startswith('#'):
                continue
            partes = line.strip().split()
            if len(partes) >= 5:
                estructura = partes[4]
                if estructura in estructuras_ventriculares:
                    volumen = float(partes[3])
                    volumen_total += volumen

    return volumen_total

# Ruta absoluta a la carpeta asegs en tu directorio actual
carpeta_asegs = '/data/Lautaro/Documentos/BrainAgeCOVID/DATOS/predictions/ventricular_volume_analyisis/asegs/asegsADNI'

# Inicializar lista para el DataFrame
data = []

# Iterar sobre todos los archivos .stats
for archivo in os.listdir(carpeta_asegs):
    if archivo.endswith('.stats'):
        ruta_archivo = os.path.join(carpeta_asegs, archivo)
        sujeto_id = archivo[:-len('aseg.stats')]  # Quitar sufijo
        volumen_total = extraer_volumenes_ventriculares(ruta_archivo)
        data.append({'ID': sujeto_id, 'Ventricles': volumen_total})

# Crear DataFrame
df_ventriculos = pd.DataFrame(data)

```

```python

def extraer_volumenes_ventriculares_csv(path_csv):
    estructuras_ventriculares = [
        'Left-Lateral-Ventricle',
        'Right-Lateral-Ventricle',
        'Left-Inf-Lat-Vent',
        'Right-Inf-Lat-Vent',
        '3rd-Ventricle',
        '4th-Ventricle',
        'Left-choroid-plexus',
        'Right-choroid-plexus'
    ]

    # Leer CSV como DataFrame
    df = pd.read_csv(path_csv)

    # Filtrar estructuras deseadas
    df_filtrado = df[df['StructName'].isin(estructuras_ventriculares)]

    # Sumar volúmenes (columna 'Volume mm3')
    volumen_total = df_filtrado['Volume_mm3'].sum()

    return volumen_total

# Ruta a la carpeta con archivos CSV de aseg
carpeta_asegs = '/data/Lautaro/Documentos/BrainAgeCOVID/DATOS/predictions/ventricular_volume_analyisis/asegs/asegsCP'

# Inicializar lista para el DataFrame
data = []

# Iterar sobre todos los archivos .csv
for archivo in os.listdir(carpeta_asegs):
    if archivo.endswith('_aseg.csv'):
        ruta_archivo = os.path.join(carpeta_asegs, archivo)
        sujeto_id = archivo.split('_')[0]  # e.g. CP0001_aseg.csv → CP0001
        volumen_total = extraer_volumenes_ventriculares_csv(ruta_archivo)
        data.append({'ID': sujeto_id, 'Ventricles': volumen_total})

# Crear DataFrame con los resultados
df_ventriculos = pd.DataFrame(data)
```
