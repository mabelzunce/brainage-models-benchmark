```python
import pandas as pd
import numpy as np
```

```python
volumCP = pd.read_csv('volumenes_asegCP.csv')
volumCogConVar= pd.read_csv('volumenes_asegCogConVar.csv')
volumNCN = pd.read_csv('volumenes_asegNCN.csv')
volumADNI = pd.read_csv('volumenes_asegADNI.csv')

csvcp=pd.read_csv('/data/Lautaro/Documentos/BrainAgeCOVID/DATOS/ResumenRespuestasACTUALIZADO.csv')
csvCogConVar=pd.read_csv('/data/Lautaro/Documentos/BrainAgeCOVID/DATOS/CogConVarparticipants.csv',sep='\t')
csvNCN=pd.read_csv('/data/Lautaro/Documentos/BrainAgeCOVID/DATOS/NCNparticipants.csv',sep='\t')
csvadni=pd.read_csv('/data/Lautaro/Documentos/BrainAgeCOVID/DATOS/ADNI_3_T1_3_13_2023.csv')
```

```python
#primero con CPstats
csvcp = csvcp[['ID', 'Edad','Grupo','Genero']]
csvcp.rename(columns={'ID': 'participant', 'Edad': 'Age', 'Grupo': 'dx', 'Genero': 'sex'}, inplace=True)
csvcp['dx'] = csvcp['dx'].replace({'CONTROL': 'UNSAM_CN', 'COVID': 'UNSAM_LC'})
csvcp['sex'] = csvcp['sex'].replace({'F': 'Female', 'M': 'Male'})
volumCP.rename(columns={'SubjectID': 'participant'},inplace=True)
brainchartCP = pd.merge(csvcp, volumCP, on='participant', how='inner')
brainchartCP['age_days'] = brainchartCP['Age'] * 365.245+280
brainchartCP['study']='UNSAM_LC'
brainchartCP['fs_version']=np.nan
brainchartCP['country']=np.nan
brainchartCP['run']=1
brainchartCP['session']=1
```

```python
#lo mismo con CogConVar
csvCogConVar = csvCogConVar[['participant_id', 'age', 'sex']]
csvCogConVar.rename(columns={'participant_id': 'participant', 'age': 'Age'}, inplace=True)
csvCogConVar['dx']='RRIB_CN'
csvCogConVar['sex'].replace({'m':'Male', 'f':'Female'}, inplace=True)
volumCogConVar.rename(columns={'SubjectID': 'participant'},inplace=True)
volumCogConVar['participant'] = volumCogConVar['participant'].str.replace('_T1w', '', regex=True)
brainchartCogConVar = pd.merge(csvCogConVar, volumCogConVar, on='participant', how='inner')
brainchartCogConVar['age_days'] = brainchartCogConVar['Age'] * 365.245+280
brainchartCogConVar['study']='RRIB'
brainchartCogConVar['fs_version']=np.nan
brainchartCogConVar['country']=np.nan
brainchartCogConVar['run']=1
brainchartCogConVar['session']=1
brainchartCogConVar['participant'] =  brainchartCogConVar['participant'] + '_RRIB'
```

```python
csvNCN= csvNCN[['participant_id', 'age', 'sex']]
csvNCN.rename(columns={'participant_id': 'participant', 'age': 'Age'}, inplace=True)
csvNCN['dx']='JUK_CN'
csvNCN['sex'].replace({'M':'Male', 'F':'Female'}, inplace=True)
volumNCN.rename(columns={'SubjectID': 'participant'},inplace=True)
volumNCN['participant'] = volumNCN['participant'].str.replace('_T1w', '', regex=True)
brainchartNCN = pd.merge(csvNCN, volumNCN, on='participant', how='inner')
brainchartNCN['age_days'] = brainchartNCN['Age'] * 365.245+280
brainchartNCN['study']='JUK'
brainchartNCN['fs_version']=np.nan
brainchartNCN['country']=np.nan
brainchartNCN['run']=1
brainchartNCN['session']=1
brainchartNCN['participant'] =  brainchartNCN['participant'] + '_JUK'
```

```python
csvadni=csvadni[['ID', 'Age','Group','Sex']]
csvadni.rename(columns={'ID': 'participant', 'Group': 'dx', 'Sex':'sex'}, inplace=True)
csvadni['sex'].replace({'F':'Female', 'M':'Male'}, inplace=True)
volumADNI.rename(columns={'SubjectID': 'participant'},inplace=True)
brainchartADNI = pd.merge(csvadni, volumADNI, on='participant', how='inner')
brainchartADNI['age_days'] = brainchartADNI['Age'] * 365.245+280
brainchartADNI['study']='ADNI'
brainchartADNI['fs_version']=np.nan
brainchartADNI['country']=np.nan
brainchartADNI['run']=1
brainchartADNI['session']=1
#separar los grupos de control de enfermedad
brainchartADNI['dx'] = brainchartADNI['dx'].replace({'CN': 'ADNI_CN', 'AD': 'ADNI_AD_MCI', 'MCI': 'ADNI_AD_MCI','SMC': 'ADNI_AD_MCI','EMCI': 'ADNI_AD_MCI','LMCI': 'ADNI_AD_MCI'})
```

```python
concatbrainchart = pd.concat([brainchartCP, brainchartCogConVar, brainchartNCN, brainchartADNI], ignore_index=True) 
```

```python
#guardar el CSV
concatbrainchart.to_csv('brainchart.csv', index=False)
```
