```python
import pandas as pd
import numpy as np
```

```python
# read volumes files
volumUNSAM = pd.read_csv('volumes_asegUNSAM.csv')
volumRRIB= pd.read_csv('volumes_asegRRIB.csv')
volumJUK = pd.read_csv('volumes_asegJUK.csv')
volumADNI = pd.read_csv('volumes_asegADNI.csv')

#read dataset's participants info
repo_dir='/Users/parri/OneDrive/Documentos/Beca PEFI/brainage-models-benchmark'

csvUNSAM=pd.read_csv(f'{repo_dir}/data/UNSAMparticipants.csv')
csvRRIB=pd.read_csv(f'{repo_dir}/data/RRIBparticipants.csv',sep='\t')
csvJUK=pd.read_csv(f'{repo_dir}/data/JUKparticipants.csv',sep='\t')
csvadni=pd.read_csv(f'{repo_dir}/data/ADNIparticipants.csv')
```

```python
# Create brainchart csv for UNSAM
csvUNSAM = csvUNSAM[['ID', 'Edad','Grupo','Genero']]
csvUNSAM.rename(columns={'ID': 'participant', 'Edad': 'Age', 'Grupo': 'dx', 'Genero': 'sex'}, inplace=True)
csvUNSAM['dx'] = csvUNSAM['dx'].replace({'CONTROL': 'UNSAM_CN', 'COVID': 'UNSAM_LC'})
csvUNSAM['sex'] = csvUNSAM['sex'].replace({'F': 'Female', 'M': 'Male'})
volumUNSAM.rename(columns={'SubjectID': 'participant'},inplace=True)
brainchartUNSAM = pd.merge(csvUNSAM, volumUNSAM, on='participant', how='inner')
brainchartUNSAM['age_days'] = brainchartUNSAM['Age'] * 365.245+280
brainchartUNSAM['study']='UNSAM_LC'
brainchartUNSAM['fs_version']="FS53"
brainchartUNSAM['country']="multinational"
brainchartUNSAM['run']=1
brainchartUNSAM['session']=1
```

```python
# Create brainchart csv for RRIB
csvRRIB = csvRRIB[['participant_id', 'age', 'sex']]
csvRRIB.rename(columns={'participant_id': 'participant', 'age': 'Age'}, inplace=True)
csvRRIB['dx']='RRIB_CN'
csvRRIB['sex'].replace({'m':'Male', 'f':'Female'}, inplace=True)
volumRRIB.rename(columns={'SubjectID': 'participant'},inplace=True)
volumRRIB['participant'] = volumRRIB['participant'].str.replace('_T1w', '', regex=True)
brainchartRRIB = pd.merge(csvRRIB, volumRRIB, on='participant', how='inner')
brainchartRRIB['age_days'] = brainchartRRIB['Age'] * 365.245+280
brainchartRRIB['study']='RRIB'
brainchartRRIB['fs_version']="FS53"
brainchartRRIB['country']="multinational"
brainchartRRIB['run']=1
brainchartRRIB['session']=1
brainchartRRIB['participant'] =  brainchartRRIB['participant'] + '_RRIB'
```

```python
#create brainchart csv for JUK
csvJUK= csvJUK[['participant_id', 'age', 'sex']]
csvJUK.rename(columns={'participant_id': 'participant', 'age': 'Age'}, inplace=True)
csvJUK['dx']='JUK_CN'
csvJUK['sex'].replace({'M':'Male', 'F':'Female'}, inplace=True)
volumJUK.rename(columns={'SubjectID': 'participant'},inplace=True)
volumJUK['participant'] = volumJUK['participant'].str.replace('_T1w', '', regex=True)
brainchartJUK = pd.merge(csvJUK, volumJUK, on='participant', how='inner')
brainchartJUK['age_days'] = brainchartJUK['Age'] * 365.245+280
brainchartJUK['study']='JUK'
brainchartJUK['fs_version']="FS53"
brainchartJUK['country']="multinational"
brainchartJUK['run']=1
brainchartJUK['session']=1
brainchartJUK['participant'] =  brainchartJUK['participant'] + '_JUK'
```

```python
# Create brainchart csv for ADNI
csvadni=csvadni[['ID', 'Age','Group','Sex']]
csvadni.rename(columns={'ID': 'participant', 'Group': 'dx', 'Sex':'sex'}, inplace=True)
csvadni['sex'].replace({'F':'Female', 'M':'Male'}, inplace=True)
volumADNI.rename(columns={'SubjectID': 'participant'},inplace=True)
brainchartADNI = pd.merge(csvadni, volumADNI, on='participant', how='inner')
brainchartADNI['age_days'] = brainchartADNI['Age'] * 365.245+280
brainchartADNI['study']='ADNI3'
brainchartADNI['fs_version']="FS53"
brainchartADNI['country']="multinational"
brainchartADNI['run']=1
brainchartADNI['session']=1
#separate AD and MCI from CN
brainchartADNI['dx'] = brainchartADNI['dx'].replace({'CN': 'ADNI_CN', 'AD': 'ADNI_AD_MCI', 'MCI': 'ADNI_AD_MCI','SMC': 'ADNI_AD_MCI','EMCI': 'ADNI_AD_MCI','LMCI': 'ADNI_AD_MCI'})
```

```python
#concat all braincharts
concatbrainchart = pd.concat([brainchartUNSAM, brainchartRRIB, brainchartJUK, brainchartADNI], ignore_index=True) 
concatbrainchart = concatbrainchart[['participant', 'Age', 'age_days', 'sex', 'study', 'fs_version', 'country', 'run', 'session', 'dx', 'GMV', 'WMV', 'sGMV', 'Ventricles']]
```
