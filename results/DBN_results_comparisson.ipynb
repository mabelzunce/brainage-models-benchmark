```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import plotly.express as px
import numpy as np
```

```python
DBN_synt_ADNI=pd.read_csv('../results/DeepBrainNet/DBN_ADNI.csv')
DBN_synt_UNSAM=pd.read_csv('../results/DeepBrainNet/DBN_UNSAM.csv')
DBN_synt_RRIB=pd.read_csv('../results/DeepBrainNet/DBN_RRIB.csv')
DBN_synt_JUK=pd.read_csv('../results/DeepBrainNet/DBN_JUK.csv')

DBN_bet_ADNI=pd.read_csv('../results/DeepBrainNet/DBN_ADNI_BET.csv')
DBN_bet_UNSAM=pd.read_csv('../results/DeepBrainNet/DBN_UNSAM_BET.csv')
DBN_bet_RRIB=pd.read_csv('../results/DeepBrainNet/DBN_RRIB_BET.csv')
DBN_bet_JUK=pd.read_csv('../results/DeepBrainNet/DBN_JUK_BET.csv')

DBN_pynet_ADNI=pd.read_csv('../results/DeepBrainNet/DBN_ADNI_pynet.csv')
DBN_pynet_UNSAM=pd.read_csv('../results/DeepBrainNet/DBN_UNSAM_pynet.csv')
DBN_pynet_RRIB=pd.read_csv('../results/DeepBrainNet/DBN_RRIB_pynet.csv')
DBN_pynet_JUK=pd.read_csv('../results/DeepBrainNet/DBN_JUK_pynet.csv')
```

```python
DBN_synt_ADNI=DBN_synt_ADNI[DBN_synt_ADNI['Pred_Age']<=1000]
DBN_synt_UNSAM=DBN_synt_UNSAM[DBN_synt_UNSAM['Grupo']=='CONTROL']
DBN_synt_ADNI=DBN_synt_ADNI[DBN_synt_ADNI['Group']=='CN']
DBN_bet_ADNI=DBN_bet_ADNI[DBN_bet_ADNI['Group']=='CN']
DBN_bet_UNSAM=DBN_bet_UNSAM[DBN_bet_UNSAM['Grupo']=='CONTROL']
DBN_pynet_ADNI=DBN_pynet_ADNI[DBN_pynet_ADNI['Group']=='CN']
DBN_pynet_UNSAM=DBN_pynet_UNSAM[DBN_pynet_UNSAM['Grupo']=='CONTROL']

MAE_synt_UNSAM=np.mean(DBN_synt_UNSAM['Absolute_Error'])
ME_synt_UNSAM=np.mean(DBN_synt_UNSAM['Error'])
MAE_synt_ADNI=np.mean(DBN_synt_ADNI['Absolute_error'])
ME_synt_ADNI=np.mean(DBN_synt_ADNI['Error'])
MAE_synt_RRIB=np.mean(DBN_synt_RRIB['Absolute_error'])
ME_synt_RRIB=np.mean(DBN_synt_RRIB['Error'])
MAE_synt_JUK=np.mean(DBN_synt_JUK['Absolute_error'])
ME_synt_JUK=np.mean(DBN_synt_JUK['Error'])

MAE_bet_UNSAM=np.mean(DBN_bet_UNSAM['Absolute_Error'])
ME_bet_UNSAM=np.mean(DBN_bet_UNSAM['Error'])
MAE_bet_ADNI=np.mean(DBN_bet_ADNI['Absolute_error'])
ME_bet_ADNI=np.mean(DBN_bet_ADNI['Error'])
MAE_bet_RRIB=np.mean(DBN_bet_RRIB['Absolute_error'])
ME_bet_RRIB=np.mean(DBN_bet_RRIB['Error'])
MAE_bet_JUK=np.mean(DBN_bet_JUK['Absolute_error'])
ME_bet_JUK=np.mean(DBN_bet_JUK['Error'])

MAE_pynet_UNSAM=np.mean(DBN_pynet_UNSAM['Absolute_Error'])
ME_pynet_UNSAM=np.mean(DBN_pynet_UNSAM['Error'])
MAE_pynet_ADNI=np.mean(DBN_pynet_ADNI['Absolute_error'])
ME_pynet_ADNI=np.mean(DBN_pynet_ADNI['Error'])
MAE_pynet_RRIB=np.mean(DBN_pynet_RRIB['Absolute_error'])
ME_pynet_RRIB=np.mean(DBN_pynet_RRIB['Error'])
MAE_pynet_JUK=np.mean(DBN_pynet_JUK['Absolute_error'])
ME_pynet_JUK=np.mean(DBN_pynet_JUK['Error'])
```

```python
fig, axes = plt.subplots(1, 3, figsize=(16, 4), sharex=True, sharey=True)
axes = axes.flatten()

# --- subplot 1: synthstrip ---
sns.scatterplot(data=DBN_synt_ADNI, y='Age', x='Pred_Age', ax=axes[0], color='red', label='ADNI')
sns.scatterplot(data=DBN_synt_UNSAM, y='Edad', x='Pred_Age', ax=axes[0], color='blue', label='UNSAM_LC')
sns.scatterplot(data=DBN_synt_RRIB, y='age', x='Pred_Age', ax=axes[0], color='orange', label='RRIB')
sns.scatterplot(data=DBN_synt_JUK, y='age', x='Pred_Age', ax=axes[0], color='green', label='JUK')
axes[0].plot([0, 100], [0, 100], color='black', label='Identity line')
axes[0].set_title('DeepBrainNet - synthstrip', fontsize=20)
axes[0].set_xlabel('Predicted Brain Age', fontsize=16)
axes[0].set_ylabel('Chronological Age', fontsize=16)
axes[0].legend(fontsize=14)
axes[0].grid(True)

# --- subplot 2: DeepBrainNet ---
sns.scatterplot(data=DBN_pynet_ADNI, y='Age', x='predicted_age', ax=axes[1], color='red')
sns.scatterplot(data=DBN_pynet_UNSAM, y='Edad', x='predicted_age', ax=axes[1], color='blue')
sns.scatterplot(data=DBN_pynet_JUK, y='age', x='predicted_age', ax=axes[1], color='green')
sns.scatterplot(data=DBN_pynet_RRIB, y='age', x='predicted_age', ax=axes[1], color='orange')
axes[1].plot([0, 100], [0, 100], color='black')
axes[1].set_title('DeepBrainNet - pynet', fontsize=20)
axes[1].set_xlabel('Predicted Brain Age', fontsize=16)
axes[1].set_ylabel('Chronological Age', fontsize=16)
axes[1].legend()
axes[1].grid(True)

# --- subplot 3: Pyment ---
sns.scatterplot(data=DBN_bet_ADNI, y='Age', x='Pred_Age', ax=axes[2], color='red')
sns.scatterplot(data=DBN_bet_UNSAM, y='Edad', x='Pred_Age', ax=axes[2], color='blue')
sns.scatterplot(data=DBN_bet_JUK, y='age', x='Pred_Age', ax=axes[2], color='green')
sns.scatterplot(data=DBN_bet_RRIB, y='age', x='Pred_Age', ax=axes[2], color='orange')
axes[2].plot([0, 100], [0, 100], color='black')
axes[2].set_title('DeepBrainNet - bet', fontsize=20)
axes[2].set_xlabel('Predicted Brain Age', fontsize=16)
axes[2].set_ylabel('Chronological Age', fontsize=16)
axes[2].legend()
axes[2].grid(True)


for ax in axes:
    ax.set_xlim(0, 100)
    ax.set_ylim(0, 100)
    ax.tick_params(axis='both', labelsize=14)  # Cambia 16 por el tamao que prefieras

plt.tight_layout()
plt.show()
```

```python
bet_MAE_list = []
bet_ME_list = []
synt_MAE_list = []
synt_ME_list = []
pynet_MAE_list = []
pynet_ME_list = []
```

```python
bet_UNSAM_sample = DBN_bet_UNSAM.sample(n=45)
bet_ADNI_sample = DBN_bet_ADNI.sample(n=45)
bet_RRIB_sample = DBN_bet_RRIB.sample(n=45)
bet_JUK_sample = DBN_bet_JUK.sample(n=45)

synt_UNSAM_sample = DBN_synt_UNSAM.sample(n=45)
synt_ADNI_sample = DBN_synt_ADNI.sample(n=45)
synt_RRIB_sample = DBN_synt_RRIB.sample(n=45)
synt_JUK_sample = DBN_synt_JUK.sample(n=45)

pynet_UNSAM_sample = DBN_pynet_UNSAM.sample(n=45)
pynet_ADNI_sample = DBN_pynet_ADNI.sample(n=45)
pynet_RRIB_sample = DBN_pynet_RRIB.sample(n=45)
pynet_JUK_sample = DBN_pynet_JUK.sample(n=45)
```

```python
bet_UNSAM_sample.rename(columns={'Edad': 'Age'}, inplace=True)
bet_UNSAM_sample=bet_UNSAM_sample[['ID','Absolute_Error','Error','Age']]
    
bet_ADNI_sample.rename(columns={'Absolute_error': 'Absolute_Error'}, inplace=True)
bet_ADNI_sample=bet_ADNI_sample[['ID','Absolute_Error','Error','Age']]

bet_JUK_sample.rename(columns={'participant_id': 'ID','age':'Age'}, inplace=True)
bet_JUK_sample.rename(columns={'Absolute_error': 'Absolute_Error'}, inplace=True)
bet_JUK_sample=bet_JUK_sample[['ID','Absolute_Error','Error','Age']]
bet_JUK_sample['ID']='JUK_' + bet_JUK_sample['ID']

bet_RRIB_sample.rename(columns={'participant_id': 'ID'}, inplace=True)
bet_RRIB_sample.rename(columns={'Absolute_error': 'Absolute_Error','age':'Age'}, inplace=True)
bet_RRIB_sample=bet_RRIB_sample[['ID','Absolute_Error','Error','Age']]

concat_bet = pd.concat([bet_UNSAM_sample, bet_ADNI_sample, bet_JUK_sample, bet_RRIB_sample], ignore_index=True)
```

```python
synt_UNSAM_sample.rename(columns={'Edad': 'Age'}, inplace=True)
synt_UNSAM_sample=synt_UNSAM_sample[['ID','Absolute_Error','Error','Age']]
    
synt_ADNI_sample.rename(columns={'Absolute_error': 'Absolute_Error'}, inplace=True)
synt_ADNI_sample=synt_ADNI_sample[['ID','Absolute_Error','Error','Age']]
synt_JUK_sample.rename(columns={'participant_id': 'ID','age':'Age'}, inplace=True)
synt_JUK_sample.rename(columns={'Absolute_error': 'Absolute_Error'}, inplace=True)
synt_JUK_sample=synt_JUK_sample[['ID','Absolute_Error','Error','Age']]
synt_JUK_sample['ID']='JUK_' + synt_JUK_sample['ID']
synt_RRIB_sample.rename(columns={'participant_id': 'ID'}, inplace=True)
synt_RRIB_sample.rename(columns={'Absolute_error': 'Absolute_Error','age':'Age'}, inplace=True)
synt_RRIB_sample=synt_RRIB_sample[['ID','Absolute_Error','Error','Age']]

concat_synt = pd.concat([synt_UNSAM_sample, synt_ADNI_sample, synt_JUK_sample, synt_RRIB_sample], ignore_index=True)
```

```python
pynet_UNSAM_sample.rename(columns={'Edad': 'Age'}, inplace=True)
pynet_UNSAM_sample=pynet_UNSAM_sample[['ID','Absolute_Error','Error','Age']]
    
pynet_ADNI_sample.rename(columns={'Absolute_error': 'Absolute_Error'}, inplace=True)
pynet_ADNI_sample=pynet_ADNI_sample[['ID','Absolute_Error','Error','Age']]
pynet_JUK_sample.rename(columns={'participant_id': 'ID','age':'Age'}, inplace=True)
pynet_JUK_sample.rename(columns={'Absolute_error': 'Absolute_Error'}, inplace=True)
pynet_JUK_sample=pynet_JUK_sample[['ID','Absolute_Error','Error','Age']]
pynet_JUK_sample['ID']='JUK_' + pynet_JUK_sample['ID']
pynet_RRIB_sample.rename(columns={'participant_id': 'ID'}, inplace=True)
pynet_RRIB_sample.rename(columns={'Absolute_error': 'Absolute_Error','age':'Age'}, inplace=True)
pynet_RRIB_sample=pynet_RRIB_sample[['ID','Absolute_Error','Error','Age']]

concat_pynet = pd.concat([pynet_UNSAM_sample, pynet_ADNI_sample, pynet_JUK_sample, pynet_RRIB_sample], ignore_index=True)
```

```python
#overall MAE for every model
bet_MAE=concat_bet['Absolute_Error'].mean()
synt_MAE=concat_synt['Absolute_Error'].mean()
pynet_MAE=concat_pynet['Absolute_Error'].mean()

#agregar a una lista los valores de mae
bet_MAE_list.append(bet_MAE)
synt_MAE_list.append(synt_MAE)
pynet_MAE_list.append(pynet_MAE)

#overall me for every model
bet_ME=concat_bet['Error'].mean()
synt_ME=concat_synt['Error'].mean()
pynet_ME=concat_pynet['Error'].mean()

bet_ME_list.append(bet_ME)
synt_ME_list.append(synt_ME)
pynet_ME_list.append(pynet_ME)
```

```python
print('MAE')
print(f"BET: {bet_MAE_list}")
print(f"synt: {synt_MAE_list}")
print(f"pynet: {pynet_MAE_list}")
print('ME')
print(f"BET: {bet_ME_list}")
print(f"synt: {synt_ME_list}")
print(f"pynet: {pynet_ME_list}")
```

```python
print('MAE')
print(f"BET MAE mean: {np.mean(bet_MAE_list)}")
print(f"synt MAE mean: {np.mean(synt_MAE_list)}")
print(f"pynet MAE mean: {np.mean(pynet_MAE_list)}")

print('ME')
print(f"bet ME mean: {np.mean(bet_ME_list)}")
print(f"synt ME mean: {np.mean(synt_ME_list)}")
print(f"pynet ME mean: {np.mean(pynet_ME_list)}")
```
