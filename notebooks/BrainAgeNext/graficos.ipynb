```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
ixiban=pd.read_csv('/home/lautaro/Documentos/BrainAgeNeXt/MedNeXt/IXIRESUMENBAN.csv')
adniban=pd.read_csv('/home/lautaro/Documentos/BrainAgeNeXt/MedNeXt/ADNIRESUMENBAN.csv')
cpban=pd.read_csv('/home/lautaro/Documentos/BrainAgeNeXt/MedNeXt/CPRESUMENBAN.csv')
```

```python
MAECNadni=adniban[adniban['Group']=='CN']['Absolute Error'].mean()
MAE_noCNadni=adniban[adniban['Group']!='CN']['Absolute Error'].mean()
MAEIXI=ixiban['Absolute Error'].mean()
MAECNcp=cpban[cpban['Grupo']=='CONTROL']['Absolute Error'].mean()
MAE_COVIDcp=cpban[cpban['Grupo']=='COVID']['Absolute Error'].mean()
```
